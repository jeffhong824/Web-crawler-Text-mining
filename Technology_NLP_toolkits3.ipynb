```python
# import PyPDF2
import PyPDF4
```

```python
path = 'C:/Users/Tingchun.TC.Hung/Desktop/Record/長庚LAMP+TFO/驗證/PDF 2 TEXT 正確度驗證/TFO/'
name = 'PMC238196'
```

```python
filename =  path + name + '.pdf'
```

```python
pdf = PyPDF4.PdfFileReader(open(filename, "rb"))
catch_paper_text = ''
for page in pdf.pages:
    print(page.extractText().replace('\n',' '))
    catch_paper_text += page.extractText().replace('\n',' ')
    
with open(path+name+'.txt','a',encoding='utf-8') as file :
    file.write(catch_paper_text)
```
