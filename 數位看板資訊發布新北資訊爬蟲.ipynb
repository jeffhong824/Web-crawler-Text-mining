# 停車場

* 行政區(25) 
* 速度快 
* 資訊精簡

https://www.parkinginfo.ntpc.gov.tw/parkingrealinfo/ParkingRealTimeInfo03.asp

```python
Names = ['八里區','三芝區','三重區','三峽區','土城區','中和區','五股區','平溪區','永和區','汐止區','坪林區','林口區','板橋區','金山區','泰山區','烏來區','貢寮區','淡水區','深坑區','新店區','新莊區','瑞芳區','樹林區','蘆洲區','鶯歌區']
```

```python
def parking(Name) :
    
    chrome_options=Options()
    chrome_options.add_argument('--headless')#無介面操作
    driver = webdriver.Chrome(chrome_options=chrome_options)
    driver.get('https://www.parkinginfo.ntpc.gov.tw/parkingrealinfo/ParkingRealTimeInfo03.asp')
    Names = ['八里區','三芝區','三重區','三峽區','土城區','中和區','五股區','平溪區','永和區','汐止區','坪林區','林口區','板橋區','金山區','泰山區','烏來區','貢寮區','淡水區','深坑區','新店區','新莊區','瑞芳區','樹林區','蘆洲區','鶯歌區']
    Numbers = [3,2,26,7,15,25,2,1,12,7,1,19,54,2,8,1,1,6,3,28,20,3,11,14,3]
    position = Names.index(Name)
    Buliding = Numbers[position]
    Buliding_status = [ [] for buliding in range(Buliding)]

    for 八里區 in range(Buliding):
        for each_info in range(1,7):
            Buliding_status[八里區].append(driver.find_element_by_xpath('/html/body/div[1]/div['+str(position+2)+']/div/div/div/div['+str(3+八里區)+']/div['+str(each_info)+']/div').text)
    driver.quit()
    return DataFrame(Buliding_status,columns=[Name,'停車場地址','及時剩餘車位數','總車位數','資料傳輸日期','資料傳輸時間'])
```

```python
parking('新莊區')
```

### 第二種網頁參考 
* 資訊較多

* 行政區(26)
* 少了一行政區 : 坪林區
* 多了一行政區 : 石門區 萬里區

https://tcdgis.ntpc.gov.tw/THI_NtpcAccessible/Parking.aspx

```python
Names2 = ['八里區','三芝區','三重區','三峽區','土城區','中和區','五股區','平溪區','永和區','石門區','汐止區','林口區','板橋區','金山區','泰山區','烏來區','貢寮區','淡水區','深坑區','新店區','新莊區','瑞芳區','樹林區','蘆洲區','鶯歌區']
```

```python
def parking2(Name):
    chrome_options=Options()
    chrome_options.add_argument('--headless')#無介面操作
    driver = webdriver.Chrome(chrome_options=chrome_options)
    Full = []
    for try_running in range(2,100):
        driver.get('https://tcdgis.ntpc.gov.tw/THI_NtpcAccessible/Parking.aspx?act=list&dat='+Name)
        try:
            driver.find_element_by_xpath('//*[@id="cont"]/div/div/table/tbody/tr['+str(try_running)+']/td/a').click()
            Each_ = []
            Name_p = driver.find_element_by_xpath('//*[@id="cont"]/div/div/table/tbody/tr[1]/th').text
            Each_.append(Name_p)
#             print(Name_p)
            for each_info in range(2,17,2):
                Each_.append(driver.find_element_by_xpath('//*[@id="cont"]/div/div/table/tbody/tr['+str(each_info)+']/td').text)

            Full.append(Each_)
        except:
#             print(try_running-2)
            break
    driver.quit()
    return DataFrame(Full,columns=[Name,'總車位','剩餘車位','區域','類型','地址','服務電話','收費方式','服務時間'])
```

```python
parking2('新莊區')
```

```python
Names3 = ['八里區','三芝區','三重區','三峽區','土城區','中和區','五股區','永和區','石門區','汐止區','坪林區','林口區','板橋區','金山區','泰山區','淡水區','深坑區','新店區','新莊區','瑞芳區','萬里區','樹林區','蘆洲區','鶯歌區']
```

```python
def Youbike(Name):
    chrome_options=Options()
    chrome_options.add_argument('--headless')#無介面操作
    driver = webdriver.Chrome(chrome_options=chrome_options)
    Full = []
    for try_running in range(2,100):
        driver.get('https://tcdgis.ntpc.gov.tw/THI_NtpcAccessible/Bike.aspx?act=list&dat='+Name)
        try:
            driver.find_element_by_xpath('//*[@id="cont"]/div/div/table/tbody/tr['+str(try_running)+']/td/a').click()
            Each_ = []
            Name_p = driver.find_element_by_xpath('//*[@id="cont"]/div/div/table/tbody/tr[1]/th').text
            Each_.append(Name_p)
#             print(Name_p)
            for each_info in range(2,7,2):
                Each_.append(driver.find_element_by_xpath('//*[@id="cont"]/div/div/table/tbody/tr['+str(each_info)+']/td').text)

            Full.append(Each_)
        except:
#             print(try_running-2)
            break
    driver.quit()
    return DataFrame(Full,columns=[Name,'地址','可停車位','可借車輛'])
```

```python
Youbike('新莊區')
```

```python
def Town_weather(Town_name):
    chrome_options=Options()
    chrome_options.add_argument('--headless')#無介面操作
    driver = webdriver.Chrome(chrome_options=chrome_options)
    Names4 = ['板橋區','三重區', '中和區', '永和區', '新莊區', '新店區', '樹林區', '鶯歌區', '三峽區', '淡水區', '汐止區', '瑞芳區', '土城區', '蘆洲區',
    '五股區', '泰山區', '林口區', '深坑區', '石碇區', '坪林區', '三芝區', '石門區', '八里區', '平溪區', '雙溪區', '貢寮區', '金山區', '萬里區', '烏來區']
    Order = Names4.index(Town_name)
    driver.get('https://www.cwb.gov.tw/V8/C/W/Town/Town.html?TID='+str(6500100+Order*100))
    Town = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[1]/div[1]/div/h2").text
    Info1 = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[2]/div[1]/div[2]/div/div/div[1]/div/div[3]").text.replace('\n','')
    Info2 = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[2]/div[1]/div[2]/div/div/div[2]/div/div[1]").text.replace('\n','') + '°C'
    Info3 = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[2]/div[1]/div[2]/div/div/div[2]/div/div[2]").text.replace('\n','') + '°C'
    Info4 = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[2]/div[1]/div[2]/div/div/div[3]/div/div[1]").text.replace('\n','')
    Info5 = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[2]/div[1]/div[2]/div/div/div[3]/div/div[2]").text.replace('\n','')
    Info6 = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[2]/div[1]/div[2]/div/div/div[4]/div/div[1]").text.replace('\n','')
    Info7 = driver.find_element_by_xpath("/html/body/div[2]/main/div/div[2]/div[1]/div[2]/div/div/div[4]/div/div[2]").text.replace('\n','')
    driver.quit()
    return DataFrame([Info1,Info2,Info3,Info4,Info5,Info6,Info7],index=['資料時間','溫度','體感溫度','相對溼度','時雨量','日出','日落'],columns=[Town]).T
```
